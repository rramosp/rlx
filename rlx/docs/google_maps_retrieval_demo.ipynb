```python
import matplotlib.pyplot as plt
from rlx import geo
from rlx.utils import read_password_protected_zip
%matplotlib inline
API_key = read_password_protected_zip("google_apikey.zip")
```

## 1. create a reference

```python
g = geo.GoogleMaps_Static_Image(7.117154, -73.113612, 10, (300,300), apikey=API_key,
                                verbose=0, maptype="satellite", savedir="imgs")
g
```

### get and save img

```python
g.save()
```
